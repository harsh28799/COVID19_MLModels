```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as color
import random
import math
import datetime
```

```python
confirmed_cases=pd.read_csv("COVID-19/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv")
deaths_reported=pd.read_csv("COVID-19/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Deaths.csv")
recovered_cases=pd.read_csv("COVID-19/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Recovered.csv")
```

```python
col=confirmed_cases.keys()
col
```

```python
confirmed=confirmed_cases.loc[:,col[4]:col[-1]]
deaths=deaths_reported.loc[:,col[4]:col[-1]]
recovered=recovered_cases.loc[:,col[4]:col[-1]]
```

```python
dates=confirmed.keys()

world_cases=[]
world_death=[]
world_recovered=[]

mortality=[]

for i in dates:
    world_cases.append(confirmed[i].sum())
    world_death.append(deaths[i].sum())
    world_recovered.append(recovered[i].sum())
    mortality.append(deaths[i].sum()/confirmed[i].sum())    
```
